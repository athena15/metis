```python
step = 10
k = 99 # randomize later
```

```python
# check in steps
def find_floor (K):   
    drop = 0
    eggs = 2
    i=0
    while (i<=100):      
        i+=10
        drop +=1
        print(f"-i--{i}")
        if(i>=K):
            eggs -= 1
            print(f"egg is broken. I have {eggs} left")
            low = i - step +1
            for x in range(low,i+1):
                print(f"-x--{x}")
                drop+=1
                if(x==K):
                    return x, drop
            
    return "something went worng"        
    
```

```python
val, drop1 = find_floor(k)
print(f'val = {val} drops = {drop1}')
```

```python
drop_list =[]
val_list = []
for i in range (1,101):
    val_list.append(i)

for kk in val_list:
    num, drop_num = find_floor(kk)
    drop_list.append(drop_num)
```
