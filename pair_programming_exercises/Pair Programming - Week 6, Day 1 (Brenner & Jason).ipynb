```python
import pandas as pd
import glob
import re

from collections import defaultdict
import numpy as np
```

```python
# neg_df = pd.DataFrame
neg = []
for file in glob.glob('txt_sentoken/neg/*'):
    with open(file, 'r') as f:
        neg.append(re.sub('\W',' ',f.read()))

pos = []
for file in glob.glob('txt_sentoken/pos/*'):
    with open(file, 'r') as f:
        pos.append(re.sub('\W', ' ', f.read()))


pos_test = np.random.choice(pos, size=200, replace=False)
neg_test = np.random.choice(neg, size=200, replace=False)

pos_train = [x for x in pos if x not in pos_test]
neg_train = [x for x in neg if x not in neg_test]
```

```python
neg_words = defaultdict(int)
pos_words = defaultdict(int)

for review in neg_train:
    for word in review.split():
        neg_words[word] += 1

for review in pos_train:
    for word in review.split():
        pos_words[word] += 1
```

```python
pos_words.values()
```

```python
pos_series = pd.Series(pos_words)
neg_series = pd.Series(neg_words)
train_words = pd.concat([pos_series, neg_series], axis = 1)
train_words = train_words.rename(columns={0:'pos', 1:'neg'})
train_words.head()
```

```python
train_words.fillna(0, inplace=True)
train_words['pos_proba'] = np.log(train_words.pos / (train_words.pos + train_words.neg))
train_words['neg_proba'] = np.log(train_words.neg / (train_words.pos + train_words.neg))
train_words.head()
```
