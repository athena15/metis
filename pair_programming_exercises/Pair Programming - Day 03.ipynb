```python
import random

def guess():
    right_number = int(input('Enter a number:'))
    guess = -1
    low = 1
    high = 101
    count = 0
    
    while guess != right_number:
        
        guess = random.randint(low, high)
        
        if guess < right_number:
            
            print('Too low.')
            low = guess
            count +=1

        elif guess > right_number:

            print('Too high.')
            high = guess
            count +=1
            
    print(f'I got it right in {count} tries')
```
